## Database Queries
Basscially a code that asks for a specific piece of informaiton from the database. In large databses, sometimes you only need access a specific part of data, and query is just telling the DBMS to give you only that piece of information. For example:
* You have a database of personal information abut employees. But for some reason you want the data from employees that are within a certain range of age, or have certain number of years of employment. A query here is just to specify the features you want to get the data that meets those features.
* Another example in our data analysis which we alwayse do: Let's say from the collected experimental data for a task, I only want the trials that were successful, or the trials with a certain feature. All I do is to specify the feature I want in the code, and extract the data that meets those features.
* Some specific examples: find() in matlab is a query code, it returns data that has a specific feature.

Sometimes, the data is so complex, that a query might itself be a complex code to narrow the search down to a very specific peice of information.

# Tables and Keys

 * A table contains a bunch of columns, each column represents a feature/attribute (name, age, whatever). Each row is a unique entry for each of those features, for example, each row contains the information of a single student. Each row should be specified with a unique key. The key is either an ID, or just an index or whatever.

 * The attributes (colomns) of a given table, could themselves be some sort of an ID, referencing the relationship of that entry to another table. For example, If we have a table list of employees with their ID as key, one attribute could be the branch id in which that employee is working, where, there is a separate table that lists those branches.

 * Another interesting thing, is that there could be an attribute that specifies the relationship between the entries. For example a colomn that specifies, for each employee, the ID of their supervisor. Note, that the supervisor is themselves an entry in that table. This can extend to any other relationship.

 * Sometimes, for a table, instead of one unique key that specifies each entry, there could be a combinaiton of keys, that together, identify a unique entry, but each one independently does not address a unique entry. In the exmaple below (ignore the index), the Employee ID and the Client ID together specify the unique key, but each one separately cannot specify unique entries because they have repetitions.



````python
import pandas as pd
import numpy as np

np.random.seed(24)
df1 = pd.DataFrame({'EmployeeID': [101,105,104,105]})
df2 = pd.DataFrame({'ClientID': [401,402,403,403]})
df3 = pd.DataFrame({'totalsale':1000*np.random.rand(4)})
df = pd.concat([df1, df2,df3], axis=1)
df = df.reset_index(drop=True)
df
````

# Basics of SQL

## Query
There are Tons of data in DB, often hidden in a complex schema, the goal is to get the data we need. An exmaple:
```
SELECT employee.name, employee.age
FROM employee
WHERE employee.salary>30000
```
Ths basically asks for certain attribute information (employee name and age) from a table (employee), where an attributes meets a given requirement (salary>30000).

# Creating tables

The first thing is to create a table using SQL commands. To create one, it's good to know about different data types. Some of the important data types are: \\
* INT : integer numbers
* DECIMAL(M,N) : similar to float in python (real numbers). with M total digits with N of them as decimal
* VARCHAR(N) : string of text with length N (N characters)
* BLOB : Binary Larg OBject (??)
* DATE : date ('YYYY-MM-DD')
* TIMESTAMP : time ('YYYY-MM-DD HH:MM:SS')

These are not all the data types, but main ones. Now when creating data tables, we should specify what data types are used for each colomn.

To create table, type the command: note that it's in capital letters- it doesn't have to be, but it is best practice to have in capital just to distinguish it from normal text.

```
CREATE TABLE name();
```

Note that all commands are followed by semicolon (;). \\


A sample code would be like:

```
CREATE TABLE student(
    student_id INT PRIMARY KEY,
    name VARCHAR(20),
    major VARCHAR(20)
);
```

Here, the table `student` is created with three colomns: the `student_id`, which is of type `INT`, and is the `PRIMARY KEY`. You can also specify the primary key colomn with:

```
CREATE TABLE student(
    student_id INT,
    name VARCHAR(20),
    major VARCHAR(20),
    PRIMARY KEY(student_id)
);
```


* You can show the table visually using the command: \
`DESCRIBE student;`

* Delete a table: \\
`DROP student;`

* Modify the table: \\
  * Add a colomn:  `ALTER TABLE student ADD gpa DECIMAL(3,2);`
  * Delete a column: `ALTER TABLE student DROP gpa;`

### Inserting data into table

The command to use is: \\
`INSERT INTO tablename VALUES(x,y,z,...);`

The values should be in the order of attributes/columns. In our example: \\
```
INSERT INTO student VALUES(1, 'Jack', 'Biology');
INSERT INTO student VALUES(2, 'Kate', 'Sociol');
INSERT INTO student(student_id, name) VALUES(3, 'Clair');
INSERT INTO student VALUES(4, 'Jack', 'Biology');
INSERT INTO student VALUES(5, 'Mike', 'Comp Sci');
```

This gives the id, name and major of the first student entry.
The general form of inserting data in the table is as follows:

`INSERT INTO student(student_id, name) VALUES(xx, 'name')`

In this case, only the attributes specified in the table student will be valued. the rest will show as NULL in the table.

**Interesting point**: when running a command in SQL, it makes the change, and if you run it again it throws error saying the change is already made. Basically when a command is run once, you can't run it again (despite matlab or python). In our case, when I ran the command above once, it created the entry, and when I ran it again, it said the entry with the given key is already inserted.

To show the table with the inserted data, use the following command: \\
`SELECT * FROM student;`

This basically selects all the information (*) from the table and displays it. `SELECT` is the command for query and selecting particular information from a table.




### More advanced ways of data entry

We can give more properties to each attribute when creating a table. For exmaple:

```
CREATE TABLE student(
    student_id INT PRIMARY KEY,
    name VARCHAR(20) NOT NULL,
    major VARCHAR(20) UNIQUE
);
```

Here we specify that name should not be NULL, and major should be unique (no repetitions).
